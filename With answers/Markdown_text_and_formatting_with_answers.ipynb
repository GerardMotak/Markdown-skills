## 🏷️ Topic 1: Headings <a id='topic-1-headings'></a>

**Quick recap:** Use `#` for H1, `##` for H2, up to `######` for H6. Always keep levels sequential.

### 📝 Exercise 1.1 — Build a Document Skeleton

Create the heading structure for a report titled **"Introduction to Machine Learning"** with the following hierarchy:

- Top-level title: *Introduction to Machine Learning*
- Section: *What is Machine Learning?*
- Sub-section: *Supervised Learning*
- Sub-sub-section: *Classification*
- Section: *Applications*

### **Your answer**

<details>
<summary>✅ <strong>Reveal Answer 1.1</strong></summary>

**Raw Markdown:**
```markdown
# Introduction to Machine Learning
## What is Machine Learning?
### Supervised Learning
#### Classification
## Applications
```

**Rendered:**

# Introduction to Machine Learning
## What is Machine Learning?
### Supervised Learning
#### Classification
## Applications

**Key points:**
- H1 (`#`) is used once for the document title
- H2 (`##`) for top-level sections
- H3 (`###`) and H4 (`####`) for nested sub-sections
- Never skip levels (e.g., H1 → H3)
</details>

## ✍️ Topic 2: Text Formatting <a id='topic-2-text-formatting'></a>

**Quick recap:**
- `**bold**` → **bold**
- `*italic*` → *italic*
- `***bold italic***` → ***bold italic***
- `~~strikethrough~~` → ~~strikethrough~~
- `` `inline code` `` → `inline code`

### 📝 Exercise 2.1 — Format a Product Description

Rewrite the sentence below using proper Markdown formatting as instructed:

> *Original:* The new MacBook Pro is incredibly fast, the old model is discontinued, and the variable name is maxSpeed.

Apply these rules:
- **"MacBook Pro"** → bold
- **"incredibly fast"** → bold italic
- **"old model is discontinued"** → strikethrough
- **"maxSpeed"** → inline code


### **Your answer**

<details>
<summary>✅ <strong>Reveal Answer 2.1</strong></summary>

**Raw Markdown:**
```markdown
The new **MacBook Pro** is ***incredibly fast***, the ~~old model is discontinued~~, and the variable name is `maxSpeed`.
```

**Rendered:**

The new **MacBook Pro** is ***incredibly fast***, the ~~old model is discontinued~~, and the variable name is `maxSpeed`.

**Key points:**
- `**text**` for bold
- `***text***` for bold + italic combined
- `~~text~~` for strikethrough
- Backticks for inline code — preserves monospace and prevents interpretation
</details>

## 📋 Topic 3: Lists <a id='topic-3-lists'></a>

**Quick recap:**
- Unordered: `- item` or `* item`
- Ordered: `1. item`
- Nested: indent by 2–3 spaces
- All ordered items can use `1.` — Markdown auto-numbers

## 💻 Topic 4: Code <a id='topic-4-code'></a>

**Quick recap:**
- Inline code: `` `code` ``
- Fenced block: wrap with ` ``` ` + language tag
- Common tags: `python`, `javascript`, `bash`, `json`, `sql`, `html`

**Function to document:**
```python
def calculate_bmi(weight_kg, height_m):
    return weight_kg / (height_m ** 2)
```


### **Your answer**

def calculate_bmi(weight_kg, height_m):

## 📊 Topic 5: Tables <a id='topic-5-tables'></a>

**Quick recap:**
```markdown
| Col1 | Col2 |
|------|------|
| A    | B    |
```
- `:---` left-align, `:---:` center, `---:` right-align

## 💬 Topic 6: Blockquotes <a id='topic-6-blockquotes'></a>

**Quick recap:**
- `>` starts a blockquote
- `> >` for nested quotes
- Can contain other Markdown (bold, lists, code)

### 📝 Exercise 6.1 — Callout Boxes

Documentation often uses styled blockquotes as callout boxes. Create **three callout-style blockquotes**:

1. A `💡 Tip` about using virtual environments in Python
2. A `⚠️ Warning` about not committing `.env` files to Git
3. A `❌ Error` message saying: *AttributeError: 'NoneType' object has no attribute 'split'*

**Your answer:** *(double-click this cell and replace this line)*

### **Your answer**

<details>
<summary>✅ <strong>Reveal Answer 6.1</strong></summary>

**Raw Markdown:**
```markdown
> 💡 **Tip:** Always use a virtual environment (`venv`) to isolate project dependencies and avoid version conflicts.

> ⚠️ **Warning:** Never commit your `.env` file to Git. Add it to `.gitignore` immediately.

> ❌ **Error:** `AttributeError: 'NoneType' object has no attribute 'split'`  
> This usually means a variable you expected to be a string is `None`. Check your data source.
```

**Rendered:**

> 💡 **Tip:** Always use a virtual environment (`venv`) to isolate project dependencies and avoid version conflicts.

> ⚠️ **Warning:** Never commit your `.env` file to Git. Add it to `.gitignore` immediately.

> ❌ **Error:** `AttributeError: 'NoneType' object has no attribute 'split'`

**Key points:**
- A blank line between `>` blocks creates separate callouts
- Two spaces at end of a line inside a blockquote creates a line break
- Mix bold, inline code, and emoji freely inside blockquotes
</details>
